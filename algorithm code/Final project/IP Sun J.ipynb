```python
import numpy as np
import gurobipy as gp
from gurobipy import *
import time
```

```python
adj = np.loadtxt("p_hat300_3.txt", dtype = int)
size = adj.max()
a = np.zeros([size,size])
for e in adj:
    a[e[0] - 1][e[1] - 1] = 1
    a[e[1] - 1][e[0] - 1] = 1
```

```python
m = gp.Model("clique")

xi = {}
for i in range(size):
    xi[i] = m.addVar(vtype=GRB.BINARY,name=str(i + 1))

t = 0
c = {}
for i in range(size):
    for j in range(i + 1, size):
        if(a[i][j] == 0):
            t = t+1
            c[t] = m.addConstr(xi[i] + xi[j] <= 1, "c" + str(t))

m.setObjective(sum(xi[f] for f in xi), GRB.MAXIMIZE)
```

```python
t1 = time.perf_counter()
m.optimize()
t2 = time.perf_counter()
```

```python
print('Cardinality: %g' % m.ObjVal)
res = []
for v in m.getVars():
    if v.X == 1:
        res.append(int(v.VarName))
print("Maximum clique:", res)
print("Time for algorithm to solve(in CPU seconds):",t2 - t1)
```
